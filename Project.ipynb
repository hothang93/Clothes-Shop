## Số lần mua của khách hàng

## Tỷ lệ khách mua hàng lại 

```python
# Loại bỏ các giá trị trùng lặp trong Data Frame dựa vào cột Mã hoá đơn, Mã khách hàng, Tổng tiền hàng sau giảm giá hoá đơn
df_tong = df_donhang.drop_duplicates(subset = ['Mã hóa đơn','Mã khách hàng','Tổng tiền hàng sau giảm giá hoá đơn'])
# Tiến hành groupby theo Năm và đếm số Mã hoá đơn
df_tong_02 = df_tong.groupby(['Năm'], as_index = False)['Mã hóa đơn'].count()
# Đổi tên cột Mã hoá đơn thành Số lần mua hàng => thay đổi luôn (inplace = True)
df_tong_02.rename(columns= {'Mã hóa đơn': 'Số lần mua hàng'}, inplace = True)
# Show DataFrame kết quả
df_tong_02
```

```python
# Tiến hành groupby theo cột Năm và Mã khách hàng. Sau đó đếm số lượng Mã hoá đơn
df_soluongkhachmua = df_tong.groupby(['Năm','Mã khách hàng'], as_index = False)['Mã hóa đơn'].count()
# Lọc ra các bản ghi có số lượng Mã hoá đơn >= 2 (Tức là đã mua lại hàng hoá)
df_soluongkhachmua = df_soluongkhachmua[df_soluongkhachmua['Mã hóa đơn'] >= 2]
# Tiếp tục tiến hành groupby theo Data Frame df_soluongkhachmua theo trường Năm => Sau đó đếm theo số lượng Mã hoá đơn
df_slkm = df_soluongkhachmua.groupby(['Năm'], as_index = False)['Mã hóa đơn'].count()
# Tiến hành đổi tên cột Mã hoá đơn thành Số khách hàng mua lại, inplace = True
df_slkm.rename(columns= {'Mã hóa đơn': 'Số khách hàng mua lại'}, inplace = True)
# Show ra DataFrame kết quả
df_slkm
```

```python
# Tiến hành Merge 2 DataFrame là df_tong_02 và df_slkm
df_tong_02 = df_tong_02.merge(df_slkm)
# Xác định cột tỷ lệ khách hàng mua lại bằng cách lấy 2 cột Số khách hàng mua lại chia cho Số lần mua hàng
df_tong_02['Tỷ lệ khách hàng mua lại'] = df_tong_02['Số khách hàng mua lại'] / df_tong_02['Số lần mua hàng']
# Tiến hành làm tròn sau dấu phẩy 2 số. Sau đó đổi sang dạng str (chuỗi) => và cuối cùng ghép thêm %
df_tong_02['Tỷ lệ khách hàng mua lại'] = round((df_tong_02['Tỷ lệ khách hàng mua lại'] * 100),2).astype(str) + '%'
# In ra kết quả
df_tong_02
```
