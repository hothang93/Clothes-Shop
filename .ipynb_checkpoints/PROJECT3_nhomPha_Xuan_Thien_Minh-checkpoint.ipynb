```python
df_tong = df_donhang.drop_duplicates(subset = ['Mã hóa đơn', 'Tổng tiền hàng sau giảm giá hoá đơn'])
df_tong_01 = df_tong.groupby(['Thời gian'], as_index = False)['Tổng tiền hàng sau giảm giá hoá đơn'].sum()
df_tong_01.rename(columns= {'Tổng tiền hàng sau giảm giá hoá đơn': 'Tổng doanh số'}, inplace = True)
df_tong_01
```

2. Doanh thu trung bình của 1 Khách hàng mang lại theo thời gian

```python
df_tong = df_donhang.drop_duplicates(subset = ['Mã hóa đơn', 'Tổng tiền hàng sau giảm giá hoá đơn'])
# Tiến hành tạo ra 1 DataFrame có tên df_donhangcopy_01 là bản sao của df_donhang
df_donhangcopy_01 = df_tong.copy()
# Chuyển đổi dữ liệu cột Thời gian về dạng datetime
df_donhangcopy_01['Thời gian'] = pd.to_datetime(df_donhangcopy_01['Thời gian'])
# Chuyển đổi dạng hiển thị cột thời gian thành tháng/năm
df_donhangcopy_01['Thời gian'] = df_donhangcopy_01['Thời gian'].dt.strftime('%m-%Y')
# Thử show ra 2 dòng đầu tiên của df_donhangcopy_01
df_donhangcopy_01.head(5)
```

```python
# Tiến hành groupby theo cột Thời gian và tính tổng Thành tiền
df_tongdtthang = df_donhangcopy_01.groupby('Thời gian', as_index = False)['Thành tiền'].sum()
# Show ra 5 dòng đầu tiên của Data Frame df_tongdtthang
df_tongdtthang.head(5)
```

```python
# Tiến hành groupby df_donhangcopy_01 theo cột Thời gian và đếm xem có bao mã khách hàng không trùng lặp
df_tbkh = df_donhangcopy_01.groupby('Thời gian', as_index = False)['Mã khách hàng'].nunique()
# Tiến hành groupby df_donhangcopy_01 theo cột Thời gian và đếm xem có bao mã khách hàng
df_sodonhang = df_donhangcopy_01.groupby('Thời gian', as_index = False)['Mã khách hàng'].count()
# Đổi tên cột Mã khách hàng thành Số khách hàng
df_sodonhang.rename(columns = {'Mã khách hàng': 'Số khách hàng'}, inplace = True)
# Gán giá trị cột Số khách hàng bằng df_tbkh
df_sodonhang['Số khách hàng'] = df_tbkh
# Show ra 5 giá trị đầu tiên của df_sodonhang
df_sodonhang.head(5)
```

```python
# Tạo ra DataFrame df_doanhthutrungbinh chứa giá trị trung bình
# bằng cách lấy cột Thành tiền của df_tongdtthang chia cho cột Số khách hàng của df_sodonhang
# Round(..., 0) => Làm tròn đến hàng đơn vị
df_doanhthutrungbinh = round(df_tongdtthang['Thành tiền'] / df_sodonhang['Số khách hàng'], 0)
# Tạo ra DataFrame df_doanhthuTB_01 bằng cách GroupBy theo cột Thời gian và đếm theo mã khách hàng
df_doanhthuTB_01 = df_donhangcopy_01.groupby('Thời gian', as_index = False)['Mã khách hàng'].count()
# Đổi tên cột Mã khách hàng thành Số khách hàng
df_doanhthuTB_01.rename(columns = {'Mã khách hàng': 'Doanh số trung bình'}, inplace = True)
# Gán giá trị cột Số khách hàng bằng df_doanhthutrungbinh
df_doanhthuTB_01['Doanh số trung bình'] = df_doanhthutrungbinh
# Show ra 10 dòng kết quả đầu tiên của DataFrame df_doanhthuTB_01
df_doanhthuTB_01.head(10)
```

Tiến hành groupby và tính tổng Giá trị trả và Doanh thu thuần theo Nhóm hàng

```python
df_tylehoantra = df_hoantra.groupby(['Nhóm hàng'], as_index = False)[['Giá trị trả', 'Doanh thu thuần']].sum()
df_tylehoantra
```

7. Tỷ lệ khách mua lại theo thời gian

```python
# Loại bỏ các giá trị trùng lặp trong Data Frame dựa vào cột Mã hoá đơn, Mã khách hàng, Tổng tiền hàng sau giảm giá hoá đơn
df_tong = df_donhang.drop_duplicates(subset = ['Mã hóa đơn','Mã khách hàng','Tổng tiền hàng sau giảm giá hoá đơn'])
# Tiến hành groupby theo Năm và đếm số Mã hoá đơn
df_tong_02 = df_tong.groupby(['Năm'], as_index = False)['Mã hóa đơn'].count()
# Đổi tên cột Mã hoá đơn thành Số lần mua hàng => thay đổi luôn (inplace = True)
df_tong_02.rename(columns= {'Mã hóa đơn': 'Số lần mua hàng'}, inplace = True)
# Show DataFrame kết quả
df_tong_02
```

```python
# Tiến hành groupby theo cột Năm và Mã khách hàng. Sau đó đếm số lượng Mã hoá đơn
df_soluongkhachmua = df_tong.groupby(['Năm','Mã khách hàng'], as_index = False)['Mã hóa đơn'].count()
# Lọc ra các bản ghi có số lượng Mã hoá đơn >= 2 (Tức là đã mua lại hàng hoá)
df_soluongkhachmua = df_soluongkhachmua[df_soluongkhachmua['Mã hóa đơn'] >= 2]
# Tiếp tục tiến hành groupby theo Data Frame df_soluongkhachmua theo trường Năm => Sau đó đếm theo số lượng Mã hoá đơn
df_slkm = df_soluongkhachmua.groupby(['Năm'], as_index = False)['Mã hóa đơn'].count()
# Tiến hành đổi tên cột Mã hoá đơn thành Số khách hàng mua lại, inplace = True
df_slkm.rename(columns= {'Mã hóa đơn': 'Số khách hàng mua lại'}, inplace = True)
# Show ra DataFrame kết quả
df_slkm
```

```python
# Tiến hành Merge 2 DataFrame là df_tong_02 và df_slkm
df_tong_02 = df_tong_02.merge(df_slkm)
# Xác định cột tỷ lệ khách hàng mua lại bằng cách lấy 2 cột Số khách hàng mua lại chia cho Số lần mua hàng
df_tong_02['Tỷ lệ khách hàng mua lại'] = df_tong_02['Số khách hàng mua lại'] / df_tong_02['Số lần mua hàng']
# Tiến hành làm tròn sau dấu phẩy 2 số. Sau đó đổi sang dạng str (chuỗi) => và cuối cùng ghép thêm %
df_tong_02['Tỷ lệ khách hàng mua lại'] = round((df_tong_02['Tỷ lệ khách hàng mua lại'] * 100),2).astype(str) + '%'
# In ra kết quả
df_tong_02
```

9. Thời gian khách mua lại 1 sản phẩm

9.1. Tương quan khách hàng giữa Online và Offline

Xác định tổng doanh số của khách mua Online và Offline theo từng tháng

```python
df_thongtinkh.head(3)
```

```python
# Tiến hành groupby theo 2 trường Chi nhánh 02 và Tháng. Sau đó tính theo trường Tổng bán trừ trả hàng
df_kh_on_off = df_thongtinkh.groupby(['Chi nhánh 02', 'Tháng'], as_index = False)['Tổng bán trừ trả hàng'].sum()
# Lọc ra các giá trị Online
df_kh_on = df_kh_on_off[df_kh_on_off['Chi nhánh 02'] == 'Online']
# Tiến hành đổi tên cột Chi nhánh 02 thành Hình thức bán
df_kh_on = df_kh_on.rename(columns = {'Chi nhánh 02': 'Hình thức bán'})
# Tiến hành reset_index lại cho đẹp
df_kh_on.reset_index(inplace = True, drop = True)
# Show ra DataFrame kết quả
df_kh_on
```

```python
# Lọc ra các giá trị Offline
df_kh_off = df_kh_on_off[df_kh_on_off['Chi nhánh 02'] == 'Offline']
# Tiến hành đổi tên cột Chi nhánh 02 thành Hình thức bán
df_kh_off = df_kh_off.rename(columns = {'Chi nhánh 02': 'Hình thức bán'})
# Show ra DataFrame kết quả
df_kh_off
```

```python
# Tiến hành Merge 2 Data Frame df_kh_on và df_kh_off với nhau theo trương Tháng
df_sosanh_onoff = df_kh_on.merge(df_kh_off, left_on = 'Tháng', right_on = 'Tháng')
# Tiến hành bỏ đi 2 cột Hình thức bán_x và Hình thức bán_y, axis = 1 => loại bỏ cột
df_sosanh_onoff.drop(['Hình thức bán_x', 'Hình thức bán_y'], axis = 1, inplace = True)
# Tiến hành đổi tên cột
df_sosanh_onoff = df_sosanh_onoff.rename(columns = {'Tổng bán trừ trả hàng_x': 'Doanh thu Online',
'Tổng bán trừ trả hàng_y': 'Doanh thu Offline'})
# Tạo ra cột Chênh lệch doanh số Online và Offline
df_sosanh_onoff['Chênh lệch doanh số Online và Offline'] = df_sosanh_onoff['Doanh thu Online'] - df_sosanh_onoff['Doanh thu Offline']
# Show ra DataFrame kết quả
df_sosanh_onoff
```

13. Doanh thu và tỷ lệ hoàn theo khu vực

13.1. Tạo ra Data Frame xác định Tổng tiền hàng sau giảm giá theo Khu vực và Mã khách hàng

```python
# Loại bỏ các giá trị trùng lặp trong Data Frame dựa vào cột Mã hóa đơn, Mã khách hàng, Tổng tiền hàng sau giảm giá hóa đơn
df_tong = df_donhang.drop_duplicates(subset=['Mã hóa đơn', 'Mã khách hàng', 'Tổng tiền hàng sau giảm giá hoá đơn'])

df_dt_kv_makh = df_tong.groupby(['Khu vực (tỉnh)', 'Mã khách hàng'], as_index = False)['Tổng tiền hàng sau giảm giá hoá đơn'].sum()

df_dt_kv_makh
```

13.2. Tạo ra Data Frame xác định Tổng tiền hàng trả sau giảm giá theo mã Khách hàng

```python
# Loại bỏ các giá trị trùng lặp trong Data Frame dựa vào cột Mã hoá đơn, Mã khách hàng, Tổng tiền hàng sau giảm giá hoá đơn
df_tong_trahang = df_trahang.drop_duplicates(subset = ['Mã trả hàng','Tổng tiền hàng trả sau giảm giá'])
df_trahang_makh = df_tong_trahang.groupby(['Mã khách hàng'], as_index = False)['Tổng tiền hàng trả sau giảm giá'].sum()
df_trahang_makh
```

```python
# Tiến hành Merge dữ liệu theo cột Mã khách hàng
df_dt_kv_makh = df_dt_kv_makh.merge(df_trahang_makh, left_on = 'Mã khách hàng', right_on = 'Mã khách hàng')
df_dt_kv_makh
```

```python
# Tiến hành groupby theo trường Khu vực (tỉnh) và tính tổng theo 2 trường 'Tổng tiền hàng sau giảm giá hoá đơn', 'Tổng tiền hàng trả sau giảm giá'
df_dt_tylehoan = df_dt_kv_makh.groupby(['Khu vực (tỉnh)'], as_index = False)[['Tổng tiền hàng sau giảm giá hoá đơn', 'Tổng tiền hàng trả sau giảm giá']].sum()
# Tạo thêm cột Tỷ lệ hoàn
df_dt_tylehoan['Tỷ lệ hoàn'] = df_dt_tylehoan['Tổng tiền hàng trả sau giảm giá'] / df_dt_tylehoan['Tổng tiền hàng sau giảm giá hoá đơn']
df_dt_tylehoan['Tỷ lệ hoàn'] = round(df_dt_tylehoan['Tỷ lệ hoàn'] * 100, 2).astype(str) + '%'
# Show ra DataFrame kết quả
df_dt_tylehoan
```

5. Tìm ra 10 sản phẩm có doanh số bán tốt nhất trong từng nhóm ngành theo tháng

```python
# Nhóm các sản phẩm theo từng tháng và doanh số của chúng
df_top10_thang = df_baocaohh.groupby(['Tháng', 'Nhóm ngành', 'Tên hàng'], as_index = False)['Doanh thu'].sum()
# Sử dụng vòng lặp từ chạy từ 1 đến 12
for i in range(1, 13):
    # Mỗi lần lặp ta sẽ lọc ra tháng thứ i
    df_top10 = df_top10_thang[df_top10_thang['Tháng'] == i]
    # Tiến hành lấy ra 10 giá trị lớn nhất theo cột Doanh thu bằng hàm nlargest
    df_top10 = df_top10.nlargest(10, 'Doanh thu')
    # Tiến hành tạo ra 1 Data Frame theo Tháng và Tên hàng
    df_top10 = df_top10[['Tháng', 'Tên hàng']]
    # Tiến hành Hide đi index cho đẹp (df.style.hide_index())
    df_top10 = df_top10.style.hide_index()
    # Sử dụng lệnh display để show ra DataFrame kết quả
    display(df_top10)
```
